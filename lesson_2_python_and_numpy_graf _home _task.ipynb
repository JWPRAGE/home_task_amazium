```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
a = np.random.randint(1, 300, 20)

min_index = np.argmin(a)
max_index = np.argmax(a)
```

```python
plt.plot(a)
plt.xlabel('Індекс')
plt.ylabel('Значення')
plt.axvline(min_index, color='r', linestyle='--', label=f'Min ({a[min_index]})', linewidth=2)
plt.axvline(max_index, color='g', linestyle='--', label=f'Max ({a[max_index]})', linewidth=2)
plt.legend()
plt.grid(True)
```
