```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1, 300, 20)
a
```

```python
custom_min_index = 2
custom_max_index = 5

min_value = a[custom_min_index]
max_value = a[custom_max_index]

plt.plot(a)
plt.xlabel('')
plt.ylabel('')
plt.axvline(custom_min_index, color='r', linestyle='--', label=f'Min ({min_value})', linewidth=2)
plt.axvline(custom_max_index, color='g', linestyle='--', label=f'Max ({max_value})', linewidth=2)
plt.legend()
plt.grid(True)
```
